```python
# Recommendation System Using Real Dataset
# Dataset: MovieLens Dataset

import pandas as pd

# -----------------------------------
# Step 1: Load Dataset
# -----------------------------------

# Download dataset from:
# https://grouplens.org/datasets/movielens/

movies = pd.read_csv("movies.csv")

# -----------------------------------
# Step 2: Show Dataset
# -----------------------------------

print("===== Movie Dataset =====")
print(movies.head())

# -----------------------------------
# Step 3: Take User Input
# -----------------------------------

genre = input("\nEnter Your Favorite Genre: ")

# -----------------------------------
# Step 4: Recommendation Logic
# -----------------------------------

recommended_movies = movies[
    movies['genres'].str.contains(genre, case=False, na=False)
]

# -----------------------------------
# Step 5: Display Recommendations
# -----------------------------------

print("\nRecommended Movies:\n")

if len(recommended_movies) > 0:

    # Show first 10 recommendations
    for movie in recommended_movies['title'].head(10):
        print("✔", movie)

else:
    print("No movies found for this genre.")
```
